## 4.
Create variable `centered_price` containing a version of the `price` column with the mean price subtracted.

(Note: this 'centering' transformation is a common preprocessing step before applying various machine learning algorithms.) 

```python
# 'centering' Transformation is a common preprocessing step befoe applying various 
# machine learning alogorithms.

reviews_price_mean = reviews.price.mean()
#reviews.price.map(lambda p:p - reviews_price_mean)
```

```python
def remean_price(row):
    row.price = row.price-reviews_price_mean# price column with the mean price subtracted.
    return row

reviews.apply(remean_price, axis='columns')
```
