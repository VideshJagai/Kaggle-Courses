# 2) How many missing data points do we have?

What percentage of the values in the dataset are missing?  Your answer should be a number between 0 and 100.  (If 1/4 of the values in the dataset are missing, the answer is 25.)

```python
missing_values = sf_permits.isna().sum()
for column in sf_permits.columns:
    missing_value_Col = sf_permits[column].isna().sum()
    if(missing_value_Col>0):
        Per_Missing_Col = (missing_value_Col/sf_permits.shape[0])*100
        print(f'Column feature: {column} with NAN distribution : {Per_Missing_Col}')
```

```python
Total_Missing_Value= missing_values.sum()
Total_DS_Cel = np.product(sf_permits.shape)
percent_missing = (Total_Missing_Value/Total_DS_Cel)*100
percent_missing
```
